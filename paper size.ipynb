```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def slice_a4_paper():
    A4_WIDTH = 210
    A4_HEIGHT = 297
    
    SQUARE_SIZE = 15

    fig, ax = plt.subplots()

    # Display A4 paper
    ax.set_xlim(0, A4_WIDTH)
    ax.set_ylim(0, A4_HEIGHT)
    ax.set_aspect('equal', 'box')
    
    if SQUARE_SIZE == 30:
        ax.set_title("A4 Paper Sliced into 30x30 mm Squares")
    elif SQUARE_SIZE == 15:
        ax.set_title("A4 Paper Sliced into 15x15 mm Squares")
    else:
        ax.set_title("A4 paper slicing Test")

    # Slice the A4 paper and add squares to the plot
    for x in range(0, A4_WIDTH, SQUARE_SIZE):
        for y in range(0, A4_HEIGHT, SQUARE_SIZE):
            square = patches.Rectangle((x, y), SQUARE_SIZE, SQUARE_SIZE, edgecolor='black', facecolor='none')
            ax.add_patch(square)

    plt.gca().invert_yaxis()  # Invert y-axis for intuitive visualization (top-left origin)
    plt.show()

if __name__ == "__main__":
    slice_a4_paper()
```
